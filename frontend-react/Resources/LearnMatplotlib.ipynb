```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## tipos de plots


### Line plots

```python
x = [1, 2, 3 , 4, 5]
y = [10, 20, 25, 30, 40]
```

```python
plt.plot(x,y, 'r'); #Muestra la gráfica, el (;) es para que muestre no muestre el objeto en consola.
plt.title('Simple Line Plot'); #Le permite colocar un titulo a la gráfica.
plt.xlabel('X-axis label'); #Permite Etiquetar el eje x
plt.ylabel('Y-axis label'); #Permite Etiquetar el eje y
plt.grid(True) #Permite cuadricular la gráfica.
```

## Bar Chart

```python
months = ['Jan', 'Feb', 'Mar', 'Apr']
revenue = [100, 200, 300, 500]
```

```python
plt.figure(figsize=(10, 6)) #Perminte editar la forma de la figura, en este caso editamos el tamaño.
plt.bar(months, revenue); #Muestra el gráfico en forma de barra.
plt.title('Monthly Revenue Bar Chart'); #Le permite colocar un titulo a la gráfica.
plt.xlabel('Months'); #Permite Etiquetar el eje x
plt.ylabel('Revenue'); #Permite Etiquetar el eje y
plt.grid(True)
```

## Subplots and Legends

```python
days = [1, 2, 3, 4, 5]
original_price = [10, 20, 25, 30, 40]
predicted_price = [8, 21, 25, 28, 38]
```

```python
plt.plot(days, original_price, 'b-o' ,label='Original Price'); # 'b' se refiere al color azul y "label" es para colocarle una etiqueta a la gráfica.
plt.plot(days, predicted_price, 'r-o' ,label='Predicted Price'); # 'r' se refiere al color rojo y "label" es para colocarle una etiqueta a la gráfica.
plt.legend() #Diferencia los gráficos utilizados mostrando el "label" de cada uno.
plt.title('Plot with legends');
plt.xlabel('Weekdays');
plt.ylabel('Price');
plt.grid(True)
```

### Subplots

```python
#Plot reutilizado Original Price.
plt.subplot(3, 1, 1)
plt.plot(days, original_price, 'b-o', label='Original Price')
plt.title('Original prices over time')
plt.xlabel('Weekdays')
plt.ylabel('Price')
plt.legend();

#Plot reutilizado Predicted Price.
plt.subplot(3, 1, 2)
plt.plot(days, predicted_price, 'r-o', label='Predicted Price')
plt.title('Predicted prices over time')
plt.xlabel('Weekdays')
plt.ylabel('Price')
plt.legend();

#Diferencia entre Plot Original Price y Plot Predicted Price.
price_difference = [o - p for o, p in zip(original_price, predicted_price)]
plt.subplot(3, 1, 3)
plt.bar(days, price_difference, color='purple', label='Price Difference')
plt.title('Predicted prices over time')
plt.xlabel('Days')
plt.ylabel('Price Difference')
plt.legend();
```

### Guardando un Plot

```python
plt.plot(days, original_price, 'b-o' ,label='Original Price'); # 'b' se refiere al color azul y "label" es para colocarle una etiqueta a la gráfica.
plt.plot(days, predicted_price, 'r-o' ,label='Predicted Price'); # 'r' se refiere al color rojo y "label" es para colocarle una etiqueta a la gráfica.
plt.legend() #Diferencia los gráficos utilizados mostrando el "label" de cada uno.
plt.title('Plot with legends');
plt.xlabel('Weekdays');
plt.ylabel('Price');
plt.grid(True)
plt.savefig('example_prediction.png', dpi=300)
```
