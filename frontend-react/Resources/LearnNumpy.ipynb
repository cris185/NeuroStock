## Crear y Leer

### 1D Array

```python
arr1 = np.array([1, 2, 3, 4, 5, 7]) #Creamos el array unidimensional.
arr1 #imprimimos el array.
```

```python
type(arr1) #Función que nos devuelve el tipo de array.
```

```python
arr1.shape #Devuelve el número de elementos del array (6,) se utiliza la "," despues del cinco ya que lo devuelve como tupla y el formato lo dicta así. 
```

```python
arr1.dtype #Muestra el tipo de dato que conforma al array.
```

```python
arr1.ndim #Muestra cuantas dimensiones tiene el array.
```

### 2D Array

```python
arr2 = np.array([[1, 4, 5], [6, 7 ,4]]) #Creamos el array de dos dimensiones, lo cual es equivalente a juntar dos arrray unidimensonales (Fila y Columna)
arr2 #Imprimimos el array.
```

```python
arr2.ndim #Devuelve el número de dimensiones, en este caso son 2.
```

```python
arr2.shape #Devuelve el número de filas y columnas.
```

```python
arr2.dtype #Muestra el tipo de dato que conforma al array.
```

```python
arr2.size #Devuelve el tamaño del array.
```

### 3D Array

```python
arr3 = np.array([[[1, 4, 5], [6, 7 ,4]], [[3, 5, 6], [8,7,5]], [[1, 4, 5], [6, 7 ,4]]]) #Creamos el  array de 3 dimensiones.
arr3 #Imprimimos el array.
```

```python
arr3.ndim #Devuelve las dimensiones del array.
```

```python
arr3.shape # Devuelve el número de arrays o layers(3), el número de filas por cada array(2) y el número de columnas(3) (3, 2, 3).
```
